```python
from __future__ import division
#
import os, sys
sys.path.append(os.getcwd()+'/..')
#
import pandas as pd
import scipy.stats as stats
#
from supports._setting import nightsafari_trips_dir
from supports._setting import MINUTE

from supports.charts import histo_cumulative, x_twin_chart
from supports._setting import DInNS_PInNS, DOutNS_PInNS
%pylab inline
```

```python
Y09 = pd.read_csv('%s/%s'%(nightsafari_trips_dir, 'Y09-nightsafari-trip.csv'))
Y10 = pd.read_csv('%s/%s'%(nightsafari_trips_dir, 'Y10-nightsafari-trip.csv'))
```

```python
from datetime import datetime
Y09['hh'] = Y09.apply(lambda row:datetime.fromtimestamp(row['start-time']).hour, axis=1)
Y10['hh'] = Y10.apply(lambda row:datetime.fromtimestamp(row['start-time']).hour, axis=1)
Y09 = Y09[( 19 <= Y09['hh'])]
Y10 = Y10[( 19 <= Y10['hh'])]
#
Y09['ns-queue-time'][Y09['ns-queue-time'] < 0] = 0
Y10['ns-queue-time'][Y10['ns-queue-time'] < 0] = 0
Y09['ns-queue-time'][Y09['ns-queue-time'] > 60*60] = 60*60
Y10['ns-queue-time'][Y10['ns-queue-time'] > 60*60] = 60*60
```

```python
x1 = [ x / MINUTE for x in Y09['ns-queue-time']]
x2 = [ x / MINUTE for x in Y10['ns-queue-time']]
print stats.ttest_ind(x1, x2, equal_var = False)
histo_cumulative('', 'Minute', '', 30, [x1, x2], ['Y2009', 'Y2010'], save_fn='cum_histo_qt_ns')
```

```python
Y09_prev_in = Y09[Y09['ns-trip-mode'] == DInNS_PInNS]
Y09_prev_out = Y09[Y09['ns-trip-mode'] == DOutNS_PInNS]
Y10_prev_in = Y10[Y10['ns-trip-mode'] == DInNS_PInNS]
Y10_prev_out = Y10[Y10['ns-trip-mode'] == DOutNS_PInNS]
#
Y09_in_hourly_gb, Y10_in_hourly_gb = Y09_prev_in.groupby(['hh']), Y10_prev_in.groupby(['hh'])
Y09_out_hourly_gb, Y10_out_hourly_gb = Y09_prev_out.groupby(['hh']), Y10_prev_out.groupby(['hh'])
#
Y09_in_hourly_qt = [ x / MINUTE for x in Y09_in_hourly_gb.mean()['ns-queue-time']]
Y10_in_hourly_qt = [ x / MINUTE for x in Y10_in_hourly_gb.mean()['ns-queue-time']]
in_diff = [Y09_in_hourly_qt[i] - Y10_in_hourly_qt[i] for i in xrange(len(Y09_in_hourly_qt))]
Y09_in_hourly_num = [ x for x in Y09_in_hourly_gb.count()['ns-queue-time']]
Y10_in_hourly_num = [ x for x in Y10_in_hourly_gb.count()['ns-queue-time']]
in_num = [Y10_in_hourly_num[i] + Y09_in_hourly_num[i] for i in xrange(len(Y09_in_hourly_num))]
#
Y09_out_hourly_qt = [ x / MINUTE for x in Y09_out_hourly_gb.mean()['ns-queue-time']]
Y10_out_hourly_qt = [ x / MINUTE for x in Y10_out_hourly_gb.mean()['ns-queue-time']]
out_diff = [Y09_out_hourly_qt[i] - Y10_out_hourly_qt[i] for i in xrange(len(Y09_out_hourly_qt))]
Y09_out_hourly_num = [ x for x in Y09_out_hourly_gb.count()['ns-queue-time']]
Y10_out_hourly_num = [ x for x in Y10_out_hourly_gb.count()['ns-queue-time']]
out_num = [Y10_out_hourly_num[i] + Y09_out_hourly_num[i] for i in xrange(len(Y09_out_hourly_num))]
```

```python
x_info = ('Time slot', [19,20,21,22,23], 0)
y_info1 = ('Minute', [Y09_in_hourly_qt, Y10_in_hourly_qt, in_diff], (-5, 40), ['Y2009', 'Y2010', 'Diff.'], 'upper left')
y_info2 = ('', [in_num], (0, 30000), ['Number of night safari trips'], 'upper right') 
print stats.ttest_rel(Y09_in_hourly_qt, Y10_in_hourly_qt)
x_twin_chart((6, 6), '', x_info, y_info1, y_info2, 'time_slot_queue_time_in_ns')
```

```python
x_info = ('Time slot', [19,20,21,22,23], 0)
y_info1 = ('Minute', [Y09_out_hourly_qt, Y10_out_hourly_qt, out_diff], (-5, 40), ['Y2009', 'Y2010', 'Diff.'], 'upper left')
y_info2 = ('', [out_num], (0, 30000), ['Number of night safari trips'], 'upper right') 
print stats.ttest_rel(Y09_out_hourly_qt, Y10_out_hourly_qt)
x_twin_chart((6, 6), '', x_info, y_info1, y_info2, 'time_slot_queue_time_out_ns')
```
