```python
from __future__ import division
#
import os, sys
sys.path.append(os.getcwd()+'/..')
#
import pandas as pd
import numpy as np
import math
#
from supports._setting import individual_dir
from supports.charts import histograms
import matplotlib.pyplot as plt
%pylab inline
# pylab.rcParams['figure.figsize'] = (20, 10)
```

```python
Y09_df = pd.read_csv('%s/Y09-summary-individual.csv' % (individual_dir))
Y10_df = pd.read_csv('%s/Y10-summary-individual.csv' % (individual_dir))
```

```python
# filtering
HOUR, CENT = 60*60, 100
PROD_LIMIT = 65 / HOUR * CENT
Y09_df = Y09_df[(Y09_df['ap-prod'] < PROD_LIMIT) & (Y09_df['total-prod'] < PROD_LIMIT)]
Y10_df = Y10_df[(Y10_df['ap-prod'] < PROD_LIMIT) & (Y10_df['total-prod'] < PROD_LIMIT)]
# both years
Y09_did, Y10_did = set(Y09_df['did']), set(Y10_df['did'])
both_years_full_drivers = list(Y09_did.intersection(Y10_did))
Y09 = Y09_df[Y09_df['did'].isin(both_years_full_drivers)]
Y10 = Y10_df[Y10_df['did'].isin(both_years_full_drivers)]
```

```python
Y09_gb, Y10_gb = Y09.groupby(['did']), Y10.groupby(['did'])
#
Y09_total_prod = Y09_gb.mean()['total-prod'].to_frame('avg_total_prod').reset_index()
Y10_total_prod = Y10_gb.mean()['total-prod'].to_frame('avg_total_prod').reset_index()
Y09_td_tuples = [tuple(x) for x in Y09_total_prod.values]
Y10_td_tuples = [tuple(x) for x in Y10_total_prod.values]
#
Y09_ap_prod = Y09_gb.mean()['ap-prod'].to_frame('avg_ap_prod').reset_index()
Y10_ap_prod = Y10_gb.mean()['ap-prod'].to_frame('avg_ap_prod').reset_index()
Y09_ad_tuples = [tuple(x) for x in Y09_ap_prod.values]
Y10_ad_tuples = [tuple(x) for x in Y10_ap_prod.values]
#
Y09_did_total_ap_prod, Y10_did_total_ap_prod = {}, {}
HOUR, CENT = 60*60, 100
for did, total_prod in Y09_td_tuples:
    Y09_did_total_ap_prod[did] = [total_prod * HOUR / CENT]
for did, total_prod in Y10_td_tuples:
    Y10_did_total_ap_prod[did] = [total_prod * HOUR / CENT]
#
for did, ap_prod in Y09_ad_tuples:
    Y09_did_total_ap_prod[did].append(ap_prod * HOUR / CENT)
for did, ap_prod in Y10_ad_tuples:
    Y10_did_total_ap_prod[did].append(ap_prod * HOUR / CENT)
```

```python
Y09_x, Y09_y, Y10_x, Y10_y = [], [], [], []
#
for did, (Y09_tp, Y09_ap) in Y09_did_total_ap_prod.iteritems():
    Y09_x.append(Y09_tp)
    Y09_y.append(Y09_ap)
    #
    Y10_tp, Y10_ap = Y10_did_total_ap_prod[did]
    Y10_x.append(Y10_tp)
    Y10_y.append(Y10_ap)
```

```python
def ranking_ascending(_l):
    l = [(x, i) for i, x in enumerate(_l)]
    return [i for _, i in sorted(l)]
```

```python
Y09_tps, Y09_aps, Y10_tps, Y10_aps = [], [], [], []
#
for did, (Y09_tp, Y09_ap) in Y09_did_total_ap_prod.iteritems():
    Y09_tps.append(Y09_tp)
    Y09_aps.append(Y09_ap)
    #
    Y10_tp, Y10_ap = Y10_did_total_ap_prod[did]
    Y10_tps.append(Y10_tp)
    Y10_aps.append(Y10_ap)
```

```python
ng_tps_ng_aps, ng_tps_pg_aps, pg_tps_ng_aps, pg_tps_pg_aps = [],[],[],[]
for i in xrange(len(Y09_tps)):
    Y09_tp, Y09_ap = Y09_tps[i], Y09_aps[i]
    Y10_tp, Y10_ap = Y10_tps[i], Y10_aps[i]
    if Y10_tp - Y09_tp < 0:
        if Y10_ap - Y09_ap < 0:
            ng_tps_ng_aps.append(i)
        else:
            ng_tps_pg_aps.append(i)
    else:
        if Y10_ap - Y09_ap < 0:
            pg_tps_ng_aps.append(i)
        else:
            pg_tps_pg_aps.append(i)
```

```python
_data = [[ len(ng_tps_ng_aps),  len(ng_tps_pg_aps)],
        [ len(pg_tps_ng_aps),  len(pg_tps_pg_aps)]
       ]
row_labels, col_labels = ['negative ap prod.', 'possitive ap prod.'], ['negative general prod.', 'possitvie general prod.']
print _data
from supports.charts import bar_table
bar_table("", 'Number of trips', row_labels, col_labels, _data)
```

```python
def draw_splot(fig, pos, arrowhead_unit, _xmax, _ymax, coords):
    ax = fig.add_subplot(1,2,pos)
    sxs, sys, exs, eys = [], [], [], []
    for sx, sy, ex, ey in coords:
        sxs.append(sx)
        sys.append(sy)
        exs.append(ex)
        eys.append(ey)
        #
        plt.plot([sx, ex], [sy, ey], 'k-')
        # add arrowhead    
        dx, dy = ex - sx, ey - sy
        norm = math.sqrt(dx*dx+dy*dy)
        udx, udy = dx/norm, dy/norm
        # Rotate by angles Pi/6 and -Pi/6
        ax = udx * math.sqrt(3)/2 - udy * 1/2
        ay = udx * 1/2 + udy * math.sqrt(3)/2
        bx = udx * math.sqrt(3)/2 + udy * 1/2
        by =  -udx * 1/2 + udy * math.sqrt(3)/2
        #
        p1_x, p1_y = ex + arrowhead_unit * -ax, ey + arrowhead_unit * -ay
        p2_x, p2_y = ex + arrowhead_unit * -bx, ey + arrowhead_unit * -by
        plt.plot([ex, p1_x], [ey, p1_y], 'k-')
        plt.plot([ex, p2_x], [ey, p2_y], 'k-')
    sp1 = plt.scatter(sxs, sys, c='blue')
    sp2 = plt.scatter(exs, eys, c='red')
    plt.legend((sp1, sp2),('Y2009', 'Y2010'), scatterpoints=1, loc='lower right')
    plt.xlim(xmin=0,xmax=_xmax*1.1)
    plt.ylim(ymin=0,ymax=_ymax*1.1)
```

```python
def draw_scatter_plots(lows_id, highs_id, coords):
    low_x1, low_y1, low_x2, low_y2 = [], [], [], []    
    high_x1, high_y1, high_x2, high_y2 = [], [], [], []
    #
    _xmax0, _ymax0 = 0, 0
    for i, v in enumerate(coords):
        if i in lows_id:
            for j, l in enumerate([low_x1, low_y1, low_x2, low_y2]):
                l.append(v[j])
            _xmax1, _ymax1 = max(max(low_x1), max(low_x2)), max(max(low_y1), max(low_y2))
            if _xmax0 < _xmax1:
                _xmax0 = _xmax1
            if _ymax0 < _ymax1:
                _ymax0 = _ymax1
        elif i in highs_id:
            for j, l in enumerate([high_x1, high_y1, high_x2, high_y2]):
                l.append(v[j])            
            _xmax1, _ymax1 = max(max(high_x1), max(high_x2)), max(max(high_y1), max(high_y2))
            if _xmax0 < _xmax1:
                _xmax0 = _xmax1
            if _ymax0 < _ymax1:
                _ymax0 = _ymax1
    fig = plt.figure(figsize=(12, 6))
    draw_splot(fig, 1, 3, _xmax0, _ymax0, zip(low_x1, low_y1, low_x2, low_y2))
    draw_splot(fig, 2, 3, _xmax0, _ymax0, zip(high_x1, high_y1, high_x2, high_y2))
    #
    plt.show()
```

```python
def draw_scatter_plot(coords):
    _xmax, _ymax = 0, 0
    for x1, y1, x2, y2 in coords:
        if _xmax < max(x1, x2):
            _xmax = max(x1, x2)
        if _ymax < max(y1, y2):
            _ymax = max(y1, y2)
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(1,1,1)
    sxs, sys, exs, eys = zip(*coords)
    sp1 = plt.scatter(sxs, sys, c='blue')
    sp2 = plt.scatter(exs, eys, c='red')
    plt.legend((sp1, sp2),('Y2009', 'Y2010'), scatterpoints=1, loc='lower right')
    plt.xlim(xmin=0,xmax=_xmax*1.1)
    plt.ylim(ymin=0,ymax=_ymax*1.1)
    plt.show()
```

```python
draw_scatter_plot(zip(Y09_x, Y09_y, Y10_x, Y10_y))
```

```python
NUM = 10
order = ranking_ascending(Y10_x)
lows_id = order[:NUM]
highs_id = order[-NUM:]
draw_scatter_plots(lows_id, highs_id, zip(Y09_x, Y09_y, Y10_x, Y10_y))
```
