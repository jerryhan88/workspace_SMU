```python
df_Y2009 = pd.read_csv('%s/%s'%(aiport_trips_dir, 'airport-trip-09.csv'))
df_Y2010 = pd.read_csv('%s/%s'%(aiport_trips_dir, 'airport-trip-10.csv'))
```
