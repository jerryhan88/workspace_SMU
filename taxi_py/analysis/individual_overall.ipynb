```python
from __future__ import division
#
import os, sys
sys.path.append(os.getcwd()+'/..')
#
import pandas as pd
import numpy as np
#
from supports._setting import individual_dir
from supports.charts import histograms
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)
```

```python
Y09_df = pd.read_csv('%s/Y09-summary-individual.csv' % (individual_dir))
Y10_df = pd.read_csv('%s/Y10-summary-individual.csv' % (individual_dir))
```

```python
HOUR = 60*60
CENT = 100

Y09_ap_prod = [x * HOUR / CENT for x in list(Y09_df['ap-prod'])]
Y10_ap_prod = [x * HOUR / CENT for x in list(Y10_df['ap-prod'])]

chart_info = [
    [('Y2009', 'S$/Hour', 'Probability', 50, Y09_ap_prod), 
     ('Y2010', 'S$/Hour', 'Probability', 50, Y10_ap_prod)]
]
histograms(chart_info)
```

```python
Y09_ap_mean, Y09_ap_std = np.mean(Y09_ap_prod), np.std(Y09_ap_prod)
Y10_ap_mean, Y10_ap_std = np.mean(Y10_ap_prod), np.std(Y10_ap_prod)

Y09_ap_prod_filtered = [x for x in Y09_ap_prod if x < Y09_ap_mean + Y09_ap_std]
Y10_ap_prod_filtered = [x for x in Y10_ap_prod if x < Y10_ap_mean + Y10_ap_std]

chart_info = [
    [('Y2009', 'S$/Hour', 'Probability', 50, Y09_ap_prod_filtered), 
     ('Y2010', 'S$/Hour', 'Probability', 50, Y10_ap_prod_filtered)]
]
histograms(chart_info)
```

```python
Y09_ap_ep = [x / CENT for x in list(Y09_df['ap-eco-profit'])]
Y10_ap_ep = [x / CENT for x in list(Y10_df['ap-eco-profit'])]

chart_info = [
    [('Y2009', 'S$', 'Probability', 50, Y09_ap_ep), 
     ('Y2010', 'S$', 'Probability', 50, Y10_ap_ep)]
]
histograms(chart_info)
```

```python
Y09_ap_mean, Y09_ap_std = np.mean(Y09_ap_ep), np.std(Y09_ap_ep)
Y10_ap_mean, Y10_ap_std = np.mean(Y10_ap_ep), np.std(Y10_ap_ep)

Y09_ap_ep_filtered = [x for x in Y09_ap_ep if -200 < x <  500]
Y10_ap_ep_filtered = [x for x in Y10_ap_ep if -200 < x <  500]

chart_info = [
    [('Y2009', 'S$', 'Probability', 40, Y09_ap_ep_filtered), 
     ('Y2010', 'S$', 'Probability', 40, Y10_ap_ep_filtered)]
]
histograms(chart_info)
```

```python
Y09_total_prod = [x * HOUR / CENT for x in list(Y09_df['total-prod'])]
Y10_total_prod = [x * HOUR / CENT for x in list(Y10_df['total-prod'])]

chart_info = [
    [('Y2009', 'S$/Hour', 'Probability', 50, Y09_total_prod), 
     ('Y2010', 'S$/Hour', 'Probability', 50, Y10_total_prod)]
]
histograms(chart_info)
```

```python
Y09_total_mean, Y09_total_std = np.mean(Y09_total_prod), np.std(Y09_total_prod)
Y10_total_mean, Y10_total_std = np.mean(Y10_total_prod), np.std(Y10_total_prod)

Y09_total_prod_filtered = [x for x in Y09_total_prod if x < 50]
Y10_total_prod_filtered = [x for x in Y10_total_prod if x < 50]

chart_info = [
    [('Y2009', 'S$/Hour', 'Probability', 40, Y09_total_prod_filtered), 
     ('Y2010', 'S$/Hour', 'Probability', 40, Y10_total_prod_filtered)]
]
histograms(chart_info)
```
